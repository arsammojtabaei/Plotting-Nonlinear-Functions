```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5, 0.1)
y = 2*(x)+3
y_noise = 2 * np.random.normal(size = x.size)
ydata = y + y_noise
plt.plot(x, ydata, 'bo')
plt.plot(x, y, 'r-')
```

```python
x = np.arange(-5, 5, 0.1)
y = (x**3) + (x**2) + (x) + 3
y_noise = 20 * np.random.normal(size=x.size)
ydata= y+ y_noise
plt.plot(x, ydata, 'bo')
plt.plot(x, y, 'r-')
```

## Quadratic

```python
x = np.arange(-5, 5, 0.1)
y = np.power(x, 2)
y_noise = 2 * np.random.normal(size = x.size)
ydata = y + y_noise
plt.plot(x, ydata, 'bo')
plt.plot(x, y, 'r-')
```

## Exponential

```python
x = np.arange(-5, 5, 0.1)
y = np.exp(x)
y_noise = 10* np.random.normal(size=x.size)
ydata = y+y_noise
plt.plot(x, ydata, 'bo')
plt.plot(x, y, 'r-')
```

## Logarithmic

```python
x = np.arange(-5, 5, 0.1)
y = np.log(x)
ydata = y+ (np.random.normal(size = x.size)/2)
plt.scatter(x, ydata, color= 'blue')
plt.plot(x, y, 'r-')
```

## Sigmoidal / Logistic

```python
x = x = np.arange(-5, 5, 0.1)
y = 1-4/(1 + np.power(3, x-2))
ydata = y + (np.random.normal(size = x.size)/2)
plt.scatter(x, ydata, color='blue')
plt.plot(x, y, 'r-')
```
